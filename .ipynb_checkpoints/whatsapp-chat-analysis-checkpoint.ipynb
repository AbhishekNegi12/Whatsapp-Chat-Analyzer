## Data Analysis

```python
import re
import pandas as pd
```

```python
f = open('A2_chat.txt','r',encoding='utf-8')
```

```python
data = f.read()  # reading entire data in form of string
```

```python
print(data)
```

```python
print(type(data))
```

```python
#convert it into a pandas dataframe
```

```python
pattern = r'\[\d{2}/\d{2}/\d{2},\s\d{1,2}:\d{2}:\d{2}\s[AP]M\]'
```

```python
messages = re.split(pattern,data)[1:]
# len(messages)
# messages
```

```python
dates = re.findall(pattern,data)
# dates
```

```python
#creating pandas dataframe
```

```python
df = pd.DataFrame({'user_message':messages, 'message_date':dates})
#convert message_date type
df['message_date'] = pd.to_datetime(df['message_date'], format = '[%d/%m/%y, %I:%M:%S %p]')
df.rename(columns={'message_date': 'date'}, inplace=True)
df.head()
```

```python
df.shape
```

```python
# import re
# Separate users and messages

users = []
messages = []
for message in df['user_message']:
    entry = re.split(r'([\w\W]+?):\s', message)  # Note the 'r' prefix
    if entry[1:]:  # User name exists
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('group_notification')
        messages.append(entry[0])

df['user'] = users
df['message'] = messages
df.drop(columns=['user_message'], inplace=True)

df.head()
```

```python
# breaking down the date column
df['year'] = df['date'].dt.year
df.head()
```

```python
df['month'] = df['date'].dt.month_name()
```

```python
df['day']=df['date'].dt.day
```

```python
df['hour'] = df['date'].dt.hour
```

```python
df['minute'] = df['date'].dt.minute
```
